```python
# Read the observed/real gene expression profiles of the plotted cells.
# "geneanno.exp_noidx_renamed.csv" is similar to "geneanno.exp.csv" which is provided by the ExPecto author, but we delete the redundant fist column and label remaining columns as new names. The reason is that the original names of tissues/cells in "./resources/modellist" and "./resources/geneanno.exp.csv" in the ExPecto repository are not consistent.
observed = pd.read_csv("geneanno.exp_noidx_renamed.csv", usecols=groups)
observed
```

```python
# Add the additional "ave" column that contains the average expression level cross different cells for each gene.
# "A pseudocount was added before log transformation (0.01, except for 0.0001 for GTEx tissues (which were averaged across individuals) due to high coverage from pooling multiple samples)" cited from the paper.
for cell in set(observed.columns):
    if cell in GTEx:
        observed[cell] = observed[cell] + 0.0001
    else:
        observed[cell] = observed[cell] + 0.01
observed["ave"] = observed.mean(axis=1)
observed
```

***
# Models trained by ourselves: xgboost = v0.9, eta = 0.001

```python
# Read the predicted gene expression profiles of the plotted cells.
predicted = pd.read_csv(
    "gene_exp_predict/Exp_plotted_our_xg0.9_eta0.001.csv", usecols=groups
)

# Because the predicted values are ones after log transformation, we need to get the values before log transformation (RPKM), in order to calculate the predicted average expression level cross different cells for each gene.
predicted = np.exp(predicted)
predicted["ave"] = predicted.mean(axis=1)

# Record the seqname and type of each gene for further filtration.
seq_type = pd.read_csv(
    "gene_exp_predict/Exp_plotted_our_xg0.9_eta0.001.csv", usecols=["seqnames", "type"]
)
predicted
```

```python
# Calculate the pearson correlation between the observed and predicted values on the held out genes.
correlation = pd.DataFrame(np.zeros((96, 96)))
for i in range(95, -1, -1):
    for j in range(96):
        row = groups[95 - i]
        column = groups[j]
        combine = pd.concat(
            [
                seq_type,
                np.log(observed[row]) - np.log(observed["ave"]),
                np.log(predicted[column]) - np.log(predicted["ave"]),
            ],
            axis=1,
            ignore_index=False,
        )
        combine = combine[combine.seqnames == "chr8"]
        combine = combine[combine.type != "rRNA"]
        correlation.iloc[i, j] = combine.iloc[:, 2].corr(
            combine.iloc[:, 3], method="pearson"
        )
```

```python
correlation
```

```python
ax = sns.heatmap(correlation, center=0, xticklabels=False, yticklabels=False)
```

```python
# Calculate the pearson correlation between the observed and predicted values on the held out genes.
correlation = pd.DataFrame(np.zeros((96, 96)))
for i in range(95, -1, -1):
    for j in range(96):
        row = groups[95 - i]
        column = groups[j]
        combine = pd.concat(
            [
                seq_type,
                np.log(observed[row]) - np.log(observed["ave"]),
                np.log(predicted[column]) - np.log(predicted["ave"]),
            ],
            axis=1,
            ignore_index=False,
        )
        combine = combine[combine.seqnames == "chr8"]
        combine = combine[combine.type != "rRNA"]
        correlation.iloc[i, j] = combine.iloc[:, 2].corr(
            combine.iloc[:, 3], method="pearson"
        )

ax = sns.heatmap(correlation, center=0, xticklabels=False, yticklabels=False)
```

```python
# Calculate the spearman correlation between the observed and predicted values on the held out genes.
correlation = pd.DataFrame(np.zeros((96, 96)))
for i in range(95, -1, -1):
    for j in range(96):
        row = groups[95 - i]
        column = groups[j]
        combine = pd.concat(
            [
                seq_type,
                np.log(observed[row]) - np.log(observed["ave"]),
                np.log(predicted[column]) - np.log(predicted["ave"]),
            ],
            axis=1,
            ignore_index=False,
        )
        combine = combine[combine.seqnames == "chr8"]
        combine = combine[combine.type != "rRNA"]
        correlation.iloc[i, j] = combine.iloc[:, 2].corr(
            combine.iloc[:, 3], method="spearman"
        )

ax = sns.heatmap(correlation, center=0, xticklabels=False, yticklabels=False)
```
